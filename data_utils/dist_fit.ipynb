```python
import numpy as np
from scipy import stats
from fitter import Fitter

data = np.loadtxt('data.csv', delimiter=',')[:, 2]
nan_rows = np.isnan(data)
data = data[~nan_rows]
```

```python
f = Fitter(data, distributions=['gamma', 'rayleigh', 'chi2', 'f'], bins=250, xmax=0.2, timeout=10000)  # 创建Fitter类
f.fit()  # 调用fit函数拟合分布
f.summary()  # 输出拟合结果
```
