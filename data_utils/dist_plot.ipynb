```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib

plt.rcParams['font.sans-serif'] = ['SimHei']
matplotlib.rcParams['image.cmap'] = 'viridis'
data = np.loadtxt('data.csv', delimiter=',')
nan_rows = np.isnan(data).any(axis=1)
data = data[~nan_rows]
state = data[:, 4]
angular = data[:, 3]
speed = data[:, 2]
```

```python
indices_roam = np.where(state == 1)
angular_roam = angular[indices_roam]
speed_roam = speed[indices_roam]
indices_dwell = np.where(state == 0)
angular_dwell = angular[indices_dwell]
speed_dwell = speed[indices_dwell]

fig, (ax0, ax1, ax2) = plt.subplots(1, 3, figsize=(16, 5))
ax0.hist2d(angular, speed, bins=200, density=True)
ax0.set(xlim=(0, 300), ylim=(0, 0.3))
x_values = [0, 135]
y_values = [0, 0.3]
ax0.plot(x_values, y_values, linestyle='-', color='red')

ax1.hist2d(angular_roam, speed_roam, bins=200, density=True)
ax1.set(xlim=(0, 100), ylim=(0, 0.3))
ax2.hist2d(angular_dwell, speed_dwell, bins=200, density=True)
ax2.set(xlim=(0, 300), ylim=(0, 0.25))
ax0.set_ylabel('线速度（mm/s）')
ax1.set_xlabel('角速度（deg/s）')
plt.tight_layout()
plt.show()
```
