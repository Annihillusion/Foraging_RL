```python
import numpy as np
import matplotlib.pyplot as plt

# plt.style.use('_mpl-gallery-nogrid')
concentration_func = np.poly1d([4.642225e+05,
                                             -2.852648e+06,
                                             7.726167e+06,
                                             -1.212951e+07,
                                             1.221637e+07,
                                             -8.258561e+06,
                                             3.818097e+06,
                                             -1.213596e+06,
                                             2.665742e+05,
                                             -4.167904e+04,
                                             5.021120e+03,
                                             -4.865155e+02,
                                             2.994606e+01,
                                             1.504611e-01,
                                             4.419983e-01
                                             ])
df = np.polyder(concentration_func)
X, Y = np.meshgrid(np.linspace(-1, 1, 800), np.linspace(-1, 1, 800))
Z = np.empty((len(X), len(Y)))
for i in range(len(X)):
    for j in range(len(Y)):
        r = np.sqrt(X[i, j]**2 + Y[i, j]**2)
        if r < 1:
            Z[i, j] = df(r)
        else:
            Z[i, j] = 0
# Z = concentration_func.deriv(np.sqrt(X**2 + Y**2)/3)
fig, ax = plt.subplots()
cs = ax.imshow(Z, cmap='viridis')
ax.set_xticklabels([])
ax.set_yticklabels([])
#plt.colorbar(cs)
fig.colorbar(cs, ax=ax)
plt.show()
```

```python
plt.rcParams['font.sans-serif'] = ['SimHei']
fig, ax = plt.subplots(figsize=(8,6))
x = np.linspace(0, 1, 100)
ax.plot(x, df(x), linewidth=3)
# ax.set_ylim(-0, 4)
ax.set_xlabel('')
ax.set_ylabel('')
plt.grid()
plt.show()
```
